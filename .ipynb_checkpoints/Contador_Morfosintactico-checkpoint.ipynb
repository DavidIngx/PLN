```python
import pandas as pd
```

```python
F = open("mia07_t3_tra_Corpus-tagged-PER1.txt","r")
```

```python
df = pd.DataFrame(columns=['A'])
```

```python
for line in F:
    df = df.append({'A': line}, ignore_index=True)
```

```python
#data Formating 
cont=0
df = df[df.A.str.contains('<doc id="', na=False) == False ]
df = df[df.A.str.contains(', , Fc 0\n', na=False) == False ]
df = df[df.A.str.contains('</doc>', na=False) == False ]
df = df[df.A.str.contains('Fp ', na=False) == False ]
df = df[df.A != "\n" ]
```

```python
df = pd.DataFrame(df.A.str.split(' ',4).tolist(), columns = ['Palabra','raiz', 'etiqueta','other'])
df2 = pd.DataFrame(df.other.str.split('\n',1).tolist(), columns = ['other', "other2"])
df["other"] = df2["other"]
```

```python
#Missing
df[df.Palabra.isna()==True]
```

```python
#conteo
VS = "DA0MS0"
total=0
DA0MS0=0
NCMS000=0
AQ0MS0=0
VMIS3S0=0
VMIP3S0=0
SPS00=0
NCMP000=0
AQ0CS0=0
for x in range(456):
    if (df["etiqueta"][x] == VS):
        total += 1
        if(df["etiqueta"][x+1] == "DA0MS0" ):
            DA0MS0 += 1
        if(df["etiqueta"][x+1] == "NCMS000" ):
            NCMS000 += 1
        if(df["etiqueta"][x+1] == "AQ0MS0" ):
            AQ0MS0 += 1
        if(df["etiqueta"][x+1] == "VMIS3S0" ):
            VMIS3S0 += 1
        if(df["etiqueta"][x+1] == "VMIP3S0" ):
            VMIP3S0 += 1
        if(df["etiqueta"][x+1] == "SPS00" ):
            SPS00 += 1
        if(df["etiqueta"][x+1] == "NCMP000" ):
            NCMP000 += 1
        if(df["etiqueta"][x+1] == "AQ0CS0" ):
            AQ0CS0 += 1
            
print(VS," VS DA0MS0  = ",DA0MS0,"/",total,"=",DA0MS0/total)        
print(VS," VS NCMS000 = ",NCMS000,"/",total,"=",NCMS000/total)        
print(VS," VS NCMS000 = ",AQ0MS0,"/",total,"=",AQ0MS0/total)        
print(VS," VS AQ0MS0  = ",VMIS3S0,"/",total,"=",VMIS3S0/total)        
print(VS," VS VMIS3S0 = ",VMIP3S0,"/",total,"=",VMIP3S0/total)        
print(VS," VS SPS00   = ",SPS00,"/",total,"=",SPS00/total)        
print(VS," VS NCMP000 = ",NCMP000,"/",total,"=",NCMP000/total)        
print(VS," VS AQ0CS0 = ",AQ0CS0,"/",total,"=",AQ0CS0/total)        

```
